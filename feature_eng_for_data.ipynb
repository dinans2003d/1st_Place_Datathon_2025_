```python
df = pd.read_csv('player_df.csv')
```

```python
df = df.drop(columns=["arena.id"])
```

```python
df['battleTime'] = pd.to_datetime(df['battleTime'])
```

```python
df = df.sort_values(['playerId', 'battleTime']).reset_index(drop=True)
```

```python
df['next_battleTime'] = df.groupby('playerId')['battleTime'].shift(-1)
```

```python
df['time_until_next'] = (df['next_battleTime'] - df['battleTime']).dt.total_seconds()
```

```python
df['hours_until_next'] = df['time_until_next'] / 3600
```

```python
df['days_until_next'] = df['time_until_next'] / 86400
```

```python
df['return_bucket'] = pd.cut(
    df['time_until_next'],
    bins=[-1, 300, 3600, 86400, 604800, float('inf')],
    labels=[
        'Play Again (<5 min)',
        'Short Break (5 min - 1 hr)',
        'Session Break (1 hr - 1 day)',
        'Daily/Weekly Return (1-7 days)',
        'Long Return (>7 days)'
    ]
)
```

```python
df.head()
```

```python
df['result'] = (df['TrophyChange'] > 0).astype(int)
```

```python
df['result'].value_counts()
```

```python

# Previous result per player
df['prev_result'] = df.groupby('playerId')['result'].shift(1)
```

```python
g = df.groupby('playerId')['result']

win_series = g.apply(
    lambda x: x * (x.groupby((x != x.shift()).cumsum()).cumcount() + 1)
)
```

```python
df['win_streak'] = win_series.reset_index(level=0, drop=True)
```

```python
loss_series = g.apply(
    lambda x: (1 - x) * ((1 - x).groupby(((1 - x) != (1 - x).shift()).cumsum()).cumcount() + 1)
)
```

```python

df['loss_streak'] = loss_series.reset_index(level=0, drop=True)
```

```python
df['prev_result'] = df['prev_result'].fillna(-1).astype(int)
```

```python
df.info()
```

```python
# Previous trophies for each player (t-1)
df['prev_trophies'] = df.groupby('playerId')['StartingTrophies'].shift(1)
```

```python
# Change in trophies vs previous match
df['trophy_change_prev'] = df['StartingTrophies'] - df['prev_trophies']
```

```python
# Net trophy change over last 3 matches (t vs t-3)
df['trophy_trend_3'] = df.groupby('playerId')['StartingTrophies'].diff(3)
```

```python
df[['playerId','StartingTrophies','prev_trophies',
    'trophy_change_prev','trophy_trend_3']].head(20)
```

```python
df['matches_played'] = df.groupby('playerId').cumcount() + 1
```

```python
df['is_first_match'] = (df['matches_played'] == 1).astype(int)
```

```python
df[['playerId','matches_played','is_first_match']].head(20)
```

```python
df['hour'] = df['battleTime'].dt.hour
```

```python

df['dayofweek'] = df['battleTime'].dt.dayofweek
```

```python

df['is_weekend'] = df['dayofweek'].isin([5, 6]).astype(int)
```

```python
import plotly.graph_objects as go


BLUE = "#1f77b4"
GOLD = "#ffbf00"


win_rate = 0.5303
loss_rate = 0.2661

outcomes = ["Loss", "Win"]
rates = [loss_rate, win_rate]
colors = [GOLD, BLUE]

fig = go.Figure()


fig.add_trace(go.Bar(
    x=outcomes,
    y=[r100 for r in rates],
    marker_color=colors,
    width=0.45
))


for x, r in zip(outcomes, rates):
    fig.add_annotation(
        x=x,
        y=r100 + 1,
        text=f"{r*100:.1f}%",
        showarrow=False,
        font=dict(size=16, color="black"),
        xanchor="center",
        yanchor="bottom"
    )

fig.update_layout(
    title="Fast Return Rate by Match Outcome",
    yaxis_title="Fast Return Rate (%)",
    xaxis_title="Match Outcome",
    template="plotly_white",
    yaxis=dict(dtick=10)
)

fig.show() 
```
